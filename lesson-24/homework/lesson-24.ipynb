## Homework 1

```python
sales_data=pd.read_csv('task/sales_data.csv')
sales_data
```

```python
Z=(sales_data
   .groupby('Category')
   .agg(total_quantity_sold=pd.NamedAgg(column='Quantity', aggfunc='sum'),
        average_price_per_unit=pd.NamedAgg(column='Price', aggfunc='mean'),
        max_quantity_sold=pd.NamedAgg(column='Quantity', aggfunc='max'),
      )
)
sales_data['Sales']=sales_data['Quantity']*sales_data['Price']

top_selling_products = sales_data.groupby(["Category", "Product"])["Quantity"].sum().reset_index()
top_selling_products = top_selling_products.loc[top_selling_products.groupby("Category")["Quantity"].idxmax()]

highest_sales_day=sales_data.groupby('Date')['Sales'].sum().idxmax()

#1
print(Z)
#2
print('\nTop-Selling Products by Category:\n', top_selling_products)
#3
print('\nDate with the highest total sales is',highest_sales_day)
```

## Homework 2

```python
customer_orders=pd.read_csv('task/customer_orders.csv')
customer_orders
```

```python
#1
grouped_co=customer_orders.groupby('CustomerID')['OrderID'].agg('count')
filter=grouped_co[grouped_co<20].index
result=customer_orders[customer_orders['CustomerID'].isin(filter)]
print('customers who have made less than 20 orders: ', result['CustomerID'].unique())
```

```python
#2
grouped_co=customer_orders.groupby('CustomerID')['Price'].agg('mean')
filter=grouped_co[grouped_co>120].index
result=customer_orders[customer_orders['CustomerID'].isin(filter)]
print('Customers who have ordered products with an average price per unit greater than $120: ', result['CustomerID'].unique())
```

```python
#3
total_quantity_price=customer_orders.groupby('Product')[['Quantity','Price']].agg('sum').reset_index()
filtered=total_quantity_price[total_quantity_price['Quantity']<5]
filtered
```
