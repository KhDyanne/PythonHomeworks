```python
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('planets')
df.head()
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
sns.lineplot(x='size',y='total_bill',data=df)
```

```python
sns.histplot(df['total_bill'], bins=20, kde=True)
```

```python
sns.barplot(x='day', y='total_bill', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
```

```python
corr=df[['total_bill', 'tip', 'size']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
sns.pairplot(df)

plt.show()
```

```python
df.head()
```

```python
g=sns.FacetGreid(df, vol='day', row='smoker')
g.map(sns.scatterplot, 'total_bill','tip')
```

```python
tips=sns.load_dataset('tips')
```

```python
plt.figure(figsize=(6,4))
plt.scatter(tips['total_bill'], tips['tip'])
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.title('Maplotlib Comparison')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day', style='day')
plt.title('Seaborn Comparison')
plt.show()
```

```python
plt.boxplot([tips[tips['day']=='Sat']['total_bill'], tips[tips['day']=='Sun']['total_bill']]))
```
